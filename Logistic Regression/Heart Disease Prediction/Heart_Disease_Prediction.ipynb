```python
# Feature engineering: Add age group
data_imputed['age_group'] = pd.cut(data_imputed['age'], bins=[0, 40, 60, 100], labels=[0, 1, 2])  # 0: Young, 1: Middle, 2: Senior
```

```python
# Correlation matrix (now that all columns are numeric)
plt.figure(figsize=(10, 8))
corr = data_imputed.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1, fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Select features and target (convert num to binary: 0 vs. 1+)
features = ['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalch', 'exang', 'oldpeak', 'slope', 'dataset', 'age_group']
X = data_imputed[features]
y = (data_imputed['num'] >= 1).astype(int)  # 0 = No disease, 1 = Disease
```

```python
# Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
# 4. Logistic Regression
# Hyperparameter tuning with GridSearchCV
param_grid = {'C': [0.01, 0.1, 1, 10, 100], 'solver': ['liblinear', 'lbfgs']}
model = GridSearchCV(LogisticRegression(random_state=42), param_grid, cv=5, scoring='accuracy')
model.fit(X_train, y_train)
```

```python
# Best model
print("\nBest Parameters:", model.best_params_)
print("Best CV Accuracy:", model.best_score_)
```

```python
# Predict on test set
y_pred = model.predict(X_test)
```

```python
# 5. Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print("\nTest Accuracy:", accuracy)
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
# ROC Curve and AUC
y_prob = model.predict_proba(X_test)[:, 1]
fpr, tpr, _ = roc_curve(y_test, y_prob)
auc = roc_auc_score(y_test, y_prob)
plt.figure(figsize=(6, 4))
plt.plot(fpr, tpr, label=f'ROC Curve (AUC = {auc:.2f})')
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve for Heart Disease Prediction')
plt.legend()
plt.show()
```
